```python
import numpy as np

def norm_density(mean, variance, x):
    exponent = (- (x - mean)**2)/(2 * variance) #i separate the exponent from the coeff
    coeff = 1 / np.sqrt(2 * np.pi * variance)
    return coeff * np.exp(exponent)

from scipy import integrate 

mean_value = 2
variance_value = 1

# mean and variance are the parameters, and the a,b are the upper and bottom limits of the integration
def integration(mean, variance, a,b):
    result, _ = integrate.quad(norm_density, a, b, args=(mean, variance)) 
# quad is just the scipy thing
# args is for the mean and variance value, to fill in the a b arguements, for the mean and variance parameters
    return result

# assign values for later, the range later
mean_value = 2
variance_value = 1
lower_limit = -1
upper_limit = 1

int_result = integration(mean_value, variance_value, lower_limit,upper_limit)
print (f"The integration is approcimately {int_result}")
# the numerical integration of the normal density func, within the range I specified

mean_h = 171
variance_h= 7.12

probability = integration(mean_h, variance_h, 162, 190)
print(f"The probability is approximately {probability}")
```
