```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(f, learning_rate, starting_point, iterations, find_max=False):
    x = starting_point
    sequence = [x]
    for i in range(iterations):
        gradient = (f(x + 0.001) - f(x - 0.001)) / 0.002
        x = x + learning_rate * gradient if not find_max else x - learning_rate * gradient
        sequence.append(x)
    
    x_values = np.linspace(-2, 2, 100)
    y_values = f(x_values)
    
    plt.plot(x_values, y_values, label='f(x)')
    plt.scatter(sequence, [f(x) for x in sequence], c='r', label='gradient descent')
    plt.scatter(sequence[0], f(sequence[0]), c='g', label='Starting Point')
    plt.scatter(sequence[-1], f(sequence[-1]), c='b', label='End Point')
    plt.legend()
    plt.show()
    
    return sequence

# Test with f(x) = x^2
f_x_squared = lambda x: x**2
gradient_descent(f_x_squared, 0.1, -1, 20)

# Test with f(x) = -3x^3 + 10x^2 - 5x - 3
f_polynomial = lambda x: -3*x**3 + 10*x**2 - 5*x - 3
gradient_descent(f_polynomial, 0.01, 2, 20)

# Test with the symmetrized funny function
def funny_function(x):
    if x > 0:
        return x**x
    elif x == 0:
        return 1
    else:
        return abs(x)**abs(x)

gradient_descent(funny_function, 0.01, 1, 20)

# Test with |x|
f_abs = lambda x: abs(x)
gradient_descent(f_abs, 0.01, 1, 20)
```
