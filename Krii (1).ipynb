```python
import sympy as sp

x, y = sp.symbols('x y')

f = sp.exp(x) * sp.sin(y) + y**3

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("df/dx:")
print(df_dx)

print("df/dy:")
print(df_dy)
```

```python
g = x**2*y + x*y**2 # define func

gradient = sp.Matrix([sp.diff(g, x), sp.diff(g, y)])

gradient_value = gradient.subs({x: 1, y: -1})

magnitude = sp.sqrt(sum([i**2 for i in gradient_value]))

print("Gradient vector at (1, -1):", gradient_value)
print("Magnitude of the gradient vector:", magnitude)
```

```python
h = sp.ln(x**2 + y**2)

h_xx = sp.diff(h, x, x)
h_yy = sp.diff(h, y, y)
h_xy = sp.diff(h, x, y)

print("dh/dx^2:")
print(h_xx)

print("dh/dy^2:")
print(h_yy)

print("df/dxdy:")
print(h_xy)

if h_xy == sp.diff(h, y, x):
    print("Symmetric.")
else:
    print("Not symmetric.")
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
